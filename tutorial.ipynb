```python
# Adding and Removing Columns
# Adding columns
coffee_data_new = coffee_data.copy()
coffee_data['price'] = 2.5
coffee_data['new_price'] = np.where(coffee_data['Coffee Type'] == 'Espresso', 3.0, 5.99)
coffee_data.drop(columns=['price'], inplace=True)
coffee_data['revenue'] = coffee_data['Units Sold'] * coffee_data['new_price']
```

```python
# Rename columns
coffee_data.rename(columns={"new_price": "price"}, inplace=True)
coffee_data.head()
```

```python
bios_data_new = bios_data.copy()
bios_data_new['first_name'] = bios_data_new['name'].str.split(' ').str[0]
bios_data_new['born_datatime'] = pd.to_datetime(bios_data_new['born_date'])
bios_data_new['born_year'] = bios_data_new['born_datatime'].dt.year

bios_data_new.head()
```

```python
bios_data_new[['name', 'born_year']]
```

```python
# Grouping Data
bios_data['height_category'] = bios_data['height_cm'].apply(lambda x: 'Short' if x > 165 else ('Average' if x > 175 else 'Tall'))
bios_data.head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row ['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 and row ['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios_data['athlete_category'] = bios_data.apply(categorize_athlete, axis=1)
bios_data.head()
```
